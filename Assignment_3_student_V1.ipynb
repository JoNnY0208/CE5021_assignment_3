# Assignment 3: Class Activation Map
In this assignment we will use the [Class Activation Mapping Technique](http://cnnlocalization.csail.mit.edu/Zhou_Learning_Deep_Features_CVPR_2016_paper.pdf) to demonstrate that object specific spatial information is available in pre-trained image classification networks.

## Introduction:
The final output convolution layers of a trained image classification network contain detailed semantic information about the input image. These feature maps, although low resolution (e.g. 7 x 7) also contain spatial information about the location of the primary object in the image. In the class activation mapping technique we wish to extract this spatial information pertaining to the identified class and overlay this activation on the input image as shown in the figure below.

Plot a test image from the dataset. The image must be unnormalised and output in the correct form [H,W,C] for display by plt.imshow using `<unnormalise image function>`. (You can re-use this function later to unnormalise images that you will overlay the class activation maps onto.)

```

imagenette_dict={0:'Tench Fish',217:'English Springer',482:'Casette Player',
                491:'Chain Saw',497:'Church',566:'French Horn',
                 569:'Grabage Truck',571:'Gas Pump',574:'Golfball',
                 662:'Modem', 701:'Parachute',864:'Tow Truck', 867:'Trailer Truck'}
mean = [0.485, 0.456, 0.406]  # Example mean used for normalization
std = [0.229, 0.224, 0.225]   # Example std used for normalization

def unnormalize_image(image, mean, std):
    """
    Unnormalize a PyTorch image tensor.

    Args:
        image (torch.Tensor): The normalized image tensor of shape [C, H, W].
        mean (list or tuple): The mean used for normalization.
        std (list or tuple): The standard deviation used for normalization.

    Returns:
        torch.Tensor: The unnormalized image tensor.
    """
    # Remove the batch dimension
    image = image.squeeze(0)

    # Check if the image is in the shape [C, H, W]
    if image.dim() == 3:
        # Unnormalize the image
        for t, m, s in zip(image, mean, std):
            t.mul_(s).add_(m)
    else:
        raise ValueError("Expected image tensor of shape [C, H, W]")

    # Convert from [C, x, y] to [x, y, C]
    image = image.permute(1, 2, 0)

    return image
```

# Class Activion Mapping
You can implement the class activation mapping algorithm in the code cell below.



```
import matplotlib.pyplot as plt
import numpy as np

# Initialize Figure
fig = plt.figure(figsize=(10,10), frameon=False)

# Iterate over the dataiter to produce a few examples

# Ensure model is eval mode to disable batch normalisation during inference.
model.eval()
for i in range(10):

    ######################################################
    # Jonatan's Code here ################################
    ######################################################
    # STEP 1

    # Take 1 image from a dataloader
    img_batch, label_batch = next(iter(data_loader))
    # move the image to a device
    img_batch = img_batch.to(device)


    # STEP 2
    # Pass image through the model
    # model output two main info dictionaries:
    # - last convolutional feature maps
    # - classifier logits

    out = model(img_batch)

    # STEP 3
    # extract both features maps and logits from the output dictionary

    feat_maps = out["op_features"]        # [1, C, h, w]
    logits    = out["op_logits"]          # [1, num_classes]

    # STEP 4
    # find predicted class for the image

    curr_class = torch.argmax(logits, dim=1).cpu().detach()

    # STEP 5
    # Get fully-connected classifiers weights for image class

    fc_weights = sd["fc.weight"]          # [num_classes, C]
    # select row for the predicted class
    class_w    = fc_weights[curr_class]   # [1, C]

    # STEP 6
    # Class Activation Map
    # I am starting here with a zero 2D map of spatial size as feature maps
    cam = torch.zeros(feat_maps.shape[-2:], device=feat_maps.device)
    # for each of the channels c, I multiply the feature map by class-specific and add to a CAM. By doing so I create a weighted sum over channels.
    for c, w in enumerate(class_w[0]):
        cam += w * feat_maps[0, c, :, :]

    # STEP 7
    # Normalise CAM for visualisation
    # remove the minimum so the smallest number is 0
    cam = cam - cam.min()
    # divide by the map so largest is 1
    cam = cam / (cam.max() + 1e-8)

    # STEP 8
    # Upsample CAM to match the input image resolution
    act_map = cam.unsqueeze(0).unsqueeze(0)  # [1, 1, h, w]
    act_map_us = F.interpolate(
        act_map,
        size=(HW_trg, HW_trg),
        mode="bilinear",
        align_corners=False,
    ).squeeze(0).squeeze(0).cpu()

    # STEP 9
    # Plotting prep
    # N. Plotting
    img = img_batch.cpu()[0]

    def get_value(dictionary, key):
        try:
          return dictionary[key]
        except KeyError:
           return "Error"
    label = get_value(imagenette_dict,int(curr_class.numpy()))

    ax = fig.add_subplot(4, 4, i + 1, xticks=[], yticks=[])
    ax.title.set_text(label)
    img=unnormalize_image(img,mean,std)
    ax.imshow(img)
    plt.imshow(img)
    plt.imshow(act_map_us.detach().numpy(), cmap=plt.cm.jet, alpha=0.4)

plt.show()
```

My plan and understand of the algoriythm is as follows:
1. Get input image from a dataloader and move it to a GPU/CPU device.
2. Get the image and pass it trough the model. the model ResNet will return two things: A) last convolutional feature maps and B) logits from the final classifier.
3. Then I select the class of interest and I do this by taking predicted class by argmax over the logits for each image.
4. Then I am getting classifier weights for that selected class.
5. Next step I am taking is myltiply each feature map channel by corresponding weights and sum all channels to get a single 2D map showing me which spatial location support class - by doing this I am computeing the class Activation Map (CAP)
6. Now when I have CAP I need to normalise it by shifting it so the minimum value is 0 and only then scale to max is = to 1. By doing this it will make it easier to display as an image heatmap.
7. Next step is to upsample the image to an input image size.
8. Final step is overlay of the heatmap on the original image.
